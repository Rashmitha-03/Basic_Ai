```python
import numpy as np

data = np.array([
    [25, 50000, 1],
    [30, np.nan, 0],
    [np.nan, 45000, 1],
    [28, 52000, np.nan]
])
print(data)
```

```python
age = data[:, 0]
print(age)
salary = data[:, 1]
label = data[:, 2]
print(salary)
print(label)
```

```python
age_mean = np.nanmean(age)
salary_mean=np.nanmean(salary)
print(age_mean)
print(salary_mean)
```

```python
age[np.isnan(age)] = age_mean
salary[np.isnan(salary)] = salary_mean
```

```python
labels_non_nan = label[~np.isnan(label)]
mode = np.bincount(labels_non_nan.astype(int)).argmax()
label[np.isnan(label)] = mode
```

```python
cleaned_data = np.column_stack((age, salary, label))

print(cleaned_data)
```

```python
data = np.array([
    [25, 50000],
    [30, 60000],
    [65, 80000]
])
print(data)
```

```python
# Normalization
min_vals = data.min(axis=0)
max_vals = data.max(axis=0)
normalized = (data - min_vals) / (max_vals - min_vals)
```

```python
mean = data.mean(axis=0)
std = data.std(axis=0)
print(mean)
print(std)
standardized = (data - mean) / std

print("Normalized:\n", normalized)
print("Standardized:\n", standardized)
```

```python
categories = np.array(["cat", "dog", "cat", "bird"])

unique = np.unique(categories)

# Create mapping
mapping = {val: idx for idx, val in enumerate(unique)}

# One-hot encoding
one_hot = np.zeros((len(categories), len(unique)))

for i, val in enumerate(categories):
    one_hot[i, mapping[val]] = 1

print(one_hot)
```

```python
y = np.array([0,0,0,0,1])

# Find minority class
unique, counts = np.unique(y, return_counts=True)

minority_class = unique[np.argmin(counts)]

# Get indices
minority_indices = np.where(y == minority_class)[0]

# Oversample
oversampled_indices = np.random.choice(minority_indices, size=counts.max(), replace=True)

balanced_y = np.concatenate([y, y[oversampled_indices]])

print(balanced_y)
```
