```python
# ================================
# 1. IMPORT LIBRARIES
# ================================
import pandas as pd
import numpy as np

# ================================
# 2. CREATE DATAFRAME
# ================================
data = {
    "id": [1, 2, 3, 4, 5, 5],
    "name": ["A", "B", "C", "D", "E", "E"],
    "age": [23, 45, 31, 35, np.nan, 28],
    "salary": [50000, 60000, 55000, 65000, 70000, 70000],
    "city": ["BLR", "MUM", "DEL", "BLR", "MUM", "MUM"]
}

df = pd.DataFrame(data)

print("\n--- ORIGINAL DATA ---")
print(df)

# ================================
# 3. BASIC INSPECTION
# ================================
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())

# ================================
# 4. COLUMN OPERATIONS
# ================================
print(df["age"])
print(df[["name", "salary"]])

# Add new column
df["bonus"] = df["salary"] * 0.10

# Rename column
df.rename(columns={"salary": "income"}, inplace=True)

# ================================
# 5. ROW SELECTION
# ================================
print(df.loc[0])      # label-based
print(df.iloc[0])     # index-based

# Conditional filtering
print(df[df["age"] > 30])

# ================================
# 6. MISSING VALUES HANDLING
# ================================
print(df.isnull())

# Fill missing
df["age"].fillna(df["age"].mean(), inplace=True)

# Drop rows with missing
df.dropna(inplace=True)

# ================================
# 7. DATA TYPE CONVERSION
# ================================
df["age"] = df["age"].astype(int)

# ================================
# 8. DUPLICATES
# ================================
print(df.duplicated())
df.drop_duplicates(inplace=True)

# ================================
# 9. SORTING
# ================================
df_sorted = df.sort_values(by="age", ascending=False)

# ================================
# 10. VALUE COUNTS / UNIQUE
# ================================
print(df["city"].value_counts())
print(df["city"].unique())

# ================================
# 11. APPLY & VECTOR OPERATIONS
# ================================
df["age_plus_5"] = df["age"].apply(lambda x: x + 5)

# Vectorized
df["double_income"] = df["income"] * 2

# ================================
# 12. GROUPBY OPERATIONS
# ================================
grouped = df.groupby("city")["income"].mean()
print(grouped)

# Multiple aggregation
agg = df.groupby("city").agg({
    "income": ["mean", "max"],
    "age": "min"
})
print(agg)

# ================================
# 13. MERGE / JOIN
# ================================
df2 = pd.DataFrame({
    "id": [1,2,3,4,5],
    "dept": ["HR","IT","FIN","IT","HR"]
})

merged = pd.merge(df, df2, on="id", how="inner")
print(merged)

# ================================
# 14. CONCAT
# ================================
concat_df = pd.concat([df, df])
print(concat_df)

# ================================
# 15. PIVOT TABLE
# ================================
pivot = pd.pivot_table(df, values="income", index="city", aggfunc="mean")
print(pivot)

# ================================
# 16. MULTI-INDEX
# ================================
multi_index = df.set_index(["city", "name"])
print(multi_index)

# ================================
# 17. STACK / UNSTACK
# ================================
stacked = multi_index.stack()
print(stacked)


# ================================
# 18. STRING OPERATIONS
# ================================
df["name_upper"] = df["name"].str.upper()

# ================================
# 19. DATE-TIME OPERATIONS
# ================================
df["date"] = pd.date_range(start="2024-01-01", periods=len(df))
df["year"] = df["date"].dt.year

# ================================
# 20. OUTLIER HANDLING
# ================================
q1 = df["income"].quantile(0.25)
q3 = df["income"].quantile(0.75)
iqr = q3 - q1

df = df[(df["income"] >= q1 - 1.5*iqr) & (df["income"] <= q3 + 1.5*iqr)]

# ================================
# 21. CORRELATION
# ================================
print(df.corr(numeric_only=True))

# ================================
# 22. RESET / SET INDEX
# ================================
df.reset_index(drop=True, inplace=True)

# ================================
# 23. EXPORT DATA
# ================================
df.to_csv("output.csv", index=False)

# ================================
# 24. PERFORMANCE OPTIMIZATION
# ================================
df["city"] = df["city"].astype("category")

print("\n--- FINAL DATA ---")
print(df)
```
