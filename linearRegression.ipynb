```python
import numpy as np

# Simple dataset
X = np.array([1, 2, 3, 4, 5])   # input
y = np.array([50, 55, 65, 70, 80])  # output
```

```python
w = 0.0   # slope
b = 0.0   # intercept

learning_rate = 0.01
epochs = 1000
n = len(X)
```

```python
for epoch in range(epochs):

    # Step 1: Predictions
    y_pred = w * X + b

    # Step 2: Compute loss (MSE)
    loss = (1/n) * np.sum((y - y_pred) ** 2)

    # Step 3: Compute gradients
    dw = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    # Step 4: Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

    # Print progress
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.2f}, w = {w:.2f}, b = {b:.2f}")
```

```python
print("\nFinal Model:")
print(f"w = {w:.2f}, b = {b:.2f}")
```

```python
x_new = 6
y_new = w * x_new + b

print(f"Prediction for x=6: {y_new:.2f}")
```

```python
import numpy as np

# Dataset
X = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([50, 55, 65, 70, 80, 85, 90, 95])

# Split (75% train, 25% test)
split_index = int(0.75 * len(X))

X_train = X[:split_index]
y_train = y[:split_index]

X_test = X[split_index:]
y_test = y[split_index:]
```

```python
w = 0.0
b = 0.0

learning_rate = 0.01
epochs = 1000
n = len(X_train)

for epoch in range(epochs):
    y_pred = w * X_train + b

    dw = (-2/n) * np.sum(X_train * (y_train - y_pred))
    db = (-2/n) * np.sum(y_train - y_pred)

    w -= learning_rate * dw
    b -= learning_rate * db
```

```python
y_test_pred = w * X_test + b
```

```python
mae = np.mean(np.abs(y_test - y_test_pred))
print("MAE:", mae)
```

```python
mse = np.mean((y_test - y_test_pred) ** 2)
print("MSE:", mse)
```

```python
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```

```python
ss_total = np.sum((y_test - np.mean(y_test)) ** 2)
ss_residual = np.sum((y_test - y_test_pred) ** 2)

r2 = 1 - (ss_residual / ss_total)
print("R2 Score:", r2)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test, label="Actual")
plt.plot(X_test, y_test_pred, label="Predicted")

plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Model Evaluation")
plt.show()
```

```python
train_pred = w * X_train + b

train_error = np.mean((y_train - train_pred) ** 2)
test_error = np.mean((y_test - y_test_pred) ** 2)

print("Train Error:", train_error)
print("Test Error:", test_error)
```
